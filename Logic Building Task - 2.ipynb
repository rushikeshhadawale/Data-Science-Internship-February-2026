```python
#Problem 1: Unique Words in a Sentence

# Unique Words Identifier Program
# This program takes a sentence as input,
# identifies all unique words, and prints:
# 1. The number of unique words
# 2. The unique words

# Getting input from the user
sentence = input("Enter sentence: ")

# Converting the sentence to lowercase (optional but recommended)
sentence = sentence.lower()

# Splitting the sentence into individual words
words = sentence.split()

# Converting the list to a set to remove duplicates
unique_words = set(words)

# Printing the results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#Problem 2: Highest Salary from Employee Data

# Employee Highest Salary Finder
# This program stores employee names and salaries
# in a dictionary and finds the employee
# with the highest salary.

# Employee dictionary (Name: Salary)
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
# max() with key=employees.get compares based on salary values
highest_paid_employee = max(employees, key=employees.get)

# Get the highest salary amount
highest_salary = employees[highest_paid_employee]

# Display result
print("Highest Salary:", highest_paid_employee, "-", highest_salary)
```

```python
#Problem 3: Find Maximum and Minimum Values

# Find Maximum and Minimum Values Program
# This program finds the maximum and minimum
# values from a given list of numbers.

# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list starting from second element
for num in numbers[1:]:
    if num > maximum:
        maximum = num   # Update maximum
    if num < minimum:
        minimum = num   # Update minimum

# Display results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Problem 4: Count Products Above a Price Threshold

# Products Price Analyzer Program
# This program counts how many product prices
# are greater than 1000.

# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Counter for products above 1000
count = 0

# Loop through each price
for price in prices:
    # Check if price is greater than 1000
    if price > 1000:
        count += 1   # Increase counter

# Display result
print("Products above 1000:", count)
```

```python
#Problem 5: Calculate Attendance Percentage

# Attendance Percentage Calculator
# This program calculates attendance percentage
# based on a list containing "P" (Present)
# and "A" (Absent).

# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count total number of days
total_days = len(attendance)

# Count number of present days
present_days = 0

for status in attendance:
    if status == "P":
        present_days += 1

# Calculate attendance percentage
attendance_percentage = (present_days / total_days) * 100

# Display result
print("Attendance Percentage:", attendance_percentage)
```

```python
# Character Frequency Counter Program
# This program counts how many times each character
# appears in a given string using a dictionary.

# Input string
text = "python"

# Dictionary to store character counts
char_count = {}

# Loop through each character in the string
for char in text:
    # If character already exists in dictionary, increment count
    if char in char_count:
        char_count[char] += 1
    else:
        # If character appears first time, initialize count as 1
        char_count[char] = 1

# Display result
print(char_count)
```

```python
# Dictionary Key Existence Checker
# ----------------------------------------
# This program checks whether a specific key
# exists in a dictionary using the 'in' keyword.

# Sample dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Check if key exists in dictionary
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Average Marks Calculator Program
# This program calculates the average marks
# from a list of student marks.

# List of marks
marks = [65, 70, 80, 75]   # Example marks

# Calculate total marks
total = sum(marks)

# Calculate number of subjects
count = len(marks)

# Calculate average
average = total / count

# Display result
print("Average Marks:", average)
```
