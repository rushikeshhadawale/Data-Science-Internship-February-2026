```python
#User Login Check Program

# This program checks whether the entered
#The username and password matched the predefined credentials.
# If both match, then the login is successful.
#Otherwise, it prints invalid credentials.

# Predefined credentials - stored values
stored_username = "admin"

stored_password = "1234"
# Taking user input

username = input("Enter Username: ")
password = input("Enter Password: ")
# Check Credentials
if username == stored_username and password == stored_password:
 print("Login Successful")
else: 
 print("Invalid Credentials")
```

```python
# Pass / Fail Analyzer Program

# The program evaluates student marks through its analysis
# which determines the number of students who passed and failed.

# List of student marks
marks = [45, 78, 90, 33, 60]

# Counters for pass and fail students
pass_count = 0
fail_count = 0

# Loop through each mark
for score in marks:
    # Check if student passed
    if score >= 50:
        pass_count += 1   # Increase pass count
    else:
        fail_count += 1   # Increase fail count

# Display results
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# Simple Data Cleaner Program

# The program removes excess spaces from user names
# and converts all characters to lowercase letters.

# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
cleaned_names = []

# Loop through each name in the list
for name in names:
    # Remove leading/trailing spaces and convert to lowercase
    cleaned_name = name.strip().lower()
    
    # Add cleaned name to new list
    cleaned_names.append(cleaned_name)

# Print cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
# The program verifies message length and identifies messages that exceed 10 character length. 

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for msg in messages:
    # Calculate length of message
    length = len(msg)
    
    # Print message length
    print("Message:", msg)
    print("Length:", length)
    
    # Check if message is longer than 10 characters
    if length > 10:
        print("Flag: Message is longer than 10 characters")
    
    print("-" * 40)  # Separator for readability
```

```python
# Error Message Detector Program

# The program counts the number of times the word "ERROR"
# appears in system log messages.

# List of log messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counter for errors
error_count = 0

# Loop through each log entry
for log in logs:
    # Check if log is an error
    if log == "ERROR":
        error_count += 1   # Increase error counter

# Print total error count
print("Total ERROR messages:", error_count)
```
