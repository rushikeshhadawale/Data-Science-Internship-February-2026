```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    # Initialize variable to store total likes
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
        
    # Print total likes
    print("Total Likes:", total_likes)
    
    # Check if post is viral
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example likes data (likes from different sources/days)
likes = [200, 150, 300, 250, 150]

# Call the function
analyze_engagement(likes)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def analyze_rainfall(rainfall_list, required_level):
    
    total = 0   # Variable to store total rainfall
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total += rain
    
    # Calculate average rainfall
    average = total / len(rainfall_list)
    
    # Print average rainfall
    print("Average Rainfall:", average)
    
    # Check rainfall status
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example rainfall data (in mm)
rainfall_data = [70, 75, 80, 65, 70]

# Required rainfall level
required = 72

# Call the function
analyze_rainfall(rainfall_data, required)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def check_duplicates(usernames):
    
    unique_users = set()   # Set to store unique usernames
    duplicate_found = False
    
    # Iterate over username list
    for user in usernames:
        
        # If username is already in set, it means a duplicate exists
        if user in unique_users:
            duplicate_found = True
            break
        else:
            unique_users.add(user)
    
    # Print result based on duplicate check
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example list of usernames
users = ["user1", "user2", "admin", "user1", "guest"]

# Call the function
check_duplicates(users)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(price_list):
    
    premium_crops = []   # List to store premium crop prices
    
    # Loop through each crop price
    for price in price_list:
        
        # Check if price is greater than 2000
        if price > 2000:
            premium_crops.append(price)
    
    # Print premium crop prices
    print("Premium Crops:", premium_crops)


# Example crop prices
crop_prices = [1500, 2500, 1800, 3200, 1900]

# Call function
filter_premium_crops(crop_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker
def check_system_health(error_count):
    # Print error count
    print("Error Count:", error_count)
    # Conditional checks
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example error count
errors = 7

# Call function
check_system_health(errors)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def validate_transaction(amount):
    
    daily_limit = 50000   # Set daily transaction limit
    
    # Print transaction amount
    print("Transaction Amount:", amount)
    
    # Check transaction status
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example transaction amount
transaction_amount = 60000

# Call function
validate_transaction(transaction_amount)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    
    present_days = 0   # Count present days
    
    # Loop to count "P"
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    # Calculate attendance percentage
    percentage = (present_days / len(attendance_list)) * 100
    
    # Print attendance percentage
    print("Attendance Percentage:", percentage)
    
    # Check eligibility condition
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example attendance data
attendance = ["P", "P", "A", "P", "P"]

# Call function
check_exam_eligibility(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def calculate_bill(units):
    
    bill = 0   # Variable to store total bill
    
    # Slab 1: First 100 units
    if units <= 100:
        bill = units * 3
    
    # Slab 2: 101-200 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    
    # Slab 3: Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    # Return both bill and usage status
    return bill, status


# Example usage
units_consumed = 250

# Call function
total_bill, usage_status = calculate_bill(units_consumed)

# Display output
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
