```python
# Employee Performance Bonus Eligibility Program
# This program identifies employees with the highest
# performance score and handles ties properly.

# Employee performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find all employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display results
print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      f"(Score: {highest_score})")
```

```python
# Search Query Keyword Analysis Program
# ------------------------------------------------
# This program:
# 1. Converts input to lowercase
# 2. Removes punctuation
# 3. Counts frequency of each keyword
# 4. Displays keywords searched more than once

import string

# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
# Translate removes punctuation characters
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency using dictionary
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 5: Filter keywords searched more than once
repeated_keywords = {}

for word, count in word_count.items():
    if count > 1:
        repeated_keywords[word] = count

# Display result
print(repeated_keywords)
```

```python
# Sensor Data Validation Program
# This program:
# 1. Finds valid sensor readings (even readings)
# 2. Stores valid readings in the form of (hour_index, reading_value) tuples
# 3. Skips invalid readings (odd readings)

# Sensor readings list (index = hour, value = reading)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings
valid_readings = []

# Iterate through readings with index
for hour, value in enumerate(sensor_readings):
    # Find if reading is valid (even)
    if value % 2 == 0:
        valid_readings.append((hour, value))

# Print result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Email Domain Usage Analysis Program
# This program:
# 1. Counts how many users belong to each email domain
# 2. Calculates the percentage usage of each domain

# List of email addresses
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Total number of emails
total_users = len(emails)

# Extract domain and count occurrences
for email in emails:
    domain = email.split("@")[1]  # Get domain part
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate and display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
# Sales Spike Detection Program
# This program:
# 1. Computes the average daily sales
# 2. Identifies days with sales exceeding the average by 30%
# 3. Prints the day number and corresponding sales amount

# List of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Compute average sales
average_sales = sum(sales) / len(sales)

# Step 2: Compute the sales threshold (30% above average)
threshold = average_sales * 1.30

# Step 3: Identify and print days with sales above threshold
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```
