```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge():
    
    # List of valid recharge plans
    valid_plans = [199, 299, 399, 599]
    
    # Loop to allow user retry until valid input is entered
    while True:
        
        # Taking recharge amount input from user
        amount = int(input("Enter recharge amount: "))
        
        # Condition 1: Check if amount is less than 50
        if amount < 50:
            print("Recharge amount must be at least ₹50. Please try again.\n")
        
        # Condition 2: Check if amount is not in valid plans
        elif amount not in valid_plans:
            print("Invalid plan selected.")
            print("Available plans are: 199, 299, 399, 599. Please try again.\n")
        
        # If both conditions are satisfied
        else:
            print("Recharge successful for ₹", amount)
            break   # Exit loop when recharge is successful


# Calling the function
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System
def check_inventory(products):
    
    # The function examines all products with their corresponding stock amounts
    for product, stock in products.items():
        
        # The system generates a reorder alert when stock falls below 15 units
        if stock < 15:
            print(product, "→ Reorder Alert (Stock:", stock, ")")
        
        # The system displays both product name and stock status when inventory meets requirements
        else:
            print(product, "→ Stock OK (Stock:", stock, ")")


# The inventory dictionary contains product names and their respective stock quantities
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Flour": 18,
    "Salt": 25
}

# The function runs with inventory data as input
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System
def process_result(marks):
    
    # Initialize total marks
    total = 0
    
    # Loop to calculate total marks
    for mark in marks:
        total += mark
    
    # Calculate average
    average = total / len(marks)
    
    # Display average marks
    print("Average Marks:", average)
    
    # Check pass or fail condition
    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")
    
    # Return average (as required in hint)
    return average


# Example list of student marks
student_marks = [60, 55, 70, 40, 50]

# Calling the function
process_result(student_marks)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
# Function to calculate cab fare
def calculate_fare(distance, peak):
    
    # Base fare ₹50 + ₹12 per km
    fare = 50 + (12 * distance)
    
    # Add 25% extra if peak hour
    if peak == "yes":
        fare = fare + (fare * 0.25)
    
    # Return final fare
    return fare


# While loop to allow retry
while True:
    
    # Take distance input
    d = float(input("Enter distance (km): "))
    
    # Ask if peak hour
    p = input("Peak hour? (yes/no): ")
    
    # Call function and store result
    total = calculate_fare(d, p)
    
    # Display total fare
    print("Total Fare: ₹", total)
    
    # Ask user if they want to calculate again
    again = input("Calculate again? (yes/no): ")
    
    # Break loop if user says no
    if again != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    
    present_days = 0   # Variable to count present days
    
    # Loop to count number of "P"
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    # Calculate attendance percentage
    percentage = (present_days / len(attendance_list)) * 100
    
    print("Attendance Percentage:", percentage)
    
    # Check eligibility condition
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Call function and store result
result = check_attendance(attendance)

# Print final result
print("Status:", result)
```

```python
# Problem 6: Password Strength Checker
def check_password(password):
    
    has_digit = False          # To check if digit exists
    has_special = False        # To check if special character exists
    
    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    # Loop through each character in password
    for ch in password:
        
        # Check if character is a digit
        if ch.isdigit():
            has_digit = True
        
        # Check if character is special (@ # $)
        if ch in "@#$":
            has_special = True
    
    # Final condition check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"


# Example password
pwd = input("Enter password: ")

# Call function and print result
result = check_password(pwd)
print(result)
```
